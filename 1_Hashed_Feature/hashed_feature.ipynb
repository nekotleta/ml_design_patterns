```python
import pandas as pd
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data'
df = pd.read_csv(url, header=None, na_values='?')
df.columns = [str('column_'+str(i)) for i in range(16)]
```

```python
def hash_n(s, n):
    num = sum([ord(c) for c in s])
    return num % n
```

```python
vals = df['column_5'].value_counts().index
```

```python
pd.DataFrame(data={"column_5_values": vals, 
                    "hash3": [hash_n(c, 3) for c in vals],
                    "hash10": [hash_n(c, 10) for c in vals],
                    "hash100": [hash_n(c, 100) for c in vals]})
```
